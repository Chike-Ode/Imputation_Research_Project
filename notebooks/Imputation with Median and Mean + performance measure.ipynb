```python
#get processed data
import numpy as np
import pandas as pd
import math

lut = pd.read_csv("https://github.com/Chike-Ode/Imputation_Research_Project/blob/master/data/processed/lutein.csv?raw=true")

sug = pd.read_csv("https://github.com/Chike-Ode/Imputation_Research_Project/blob/master/data/processed/sugars.csv?raw=true")

wat=pd.read_csv("https://github.com/Chike-Ode/Imputation_Research_Project/blob/master/data/processed/water.csv?raw=true")

#adding indicators to retrieve rows that were NaN after imputation
lut['mcarna'] = np.where(np.isnan(lut['lutein_zeaxanthin_mcar'].values), 1, 0) 
lut['marna'] = np.where(np.isnan(lut['lutein_zeaxanthin_mar'].values), 1, 0) 

sug['mcarna'] = np.where(np.isnan(sug['sugars_mcar'].values), 1, 0) 
sug['marna'] = np.where(np.isnan(sug['sugars_mar'].values), 1, 0) 

wat['mcarna'] = np.where(np.isnan(wat['water_mcar'].values), 1, 0) 
wat['marna'] = np.where(np.isnan(wat['water_mar'].values), 1, 0)
```

```python
#replace NaN with median and mean for each df. Mode doesn't make much sense here with continuous data

#lutein
lutmean=lut.copy()
lutmean['lutein_zeaxanthin_mcar'].fillna(lutmean['lutein_zeaxanthin_mcar'].mean(), inplace=True)
lutmean['lutein_zeaxanthin_mar'].fillna(lutmean['lutein_zeaxanthin_mar'].mean(), inplace=True)

lutmdn=lut.copy()
lutmdn['lutein_zeaxanthin_mcar'].fillna(lutmdn['lutein_zeaxanthin_mcar'].median(), inplace=True)
lutmdn['lutein_zeaxanthin_mar'].fillna(lutmdn['lutein_zeaxanthin_mar'].median(), inplace=True)
#Median is zero here (heavily skewed to the right)

#sugars
sugmean=sug.copy()
sugmean['sugars_mcar'].fillna(sugmean['sugars_mcar'].mean(), inplace=True)
sugmean['sugars_mar'].fillna(sugmean['sugars_mar'].mean(), inplace=True)

sugmdn=sug.copy()
sugmdn['sugars_mcar'].fillna(sugmdn['sugars_mcar'].median(), inplace=True)
sugmdn['sugars_mar'].fillna(sugmdn['sugars_mar'].median(), inplace=True)

#water
watmean=wat.copy()
watmean['water_mcar'].fillna(watmean['water_mcar'].mean(), inplace=True)
watmean['water_mar'].fillna(watmean['water_mar'].mean(), inplace=True)

watmdn=wat.copy()
watmdn['water_mcar'].fillna(watmdn['water_mcar'].median(), inplace=True)
watmdn['water_mar'].fillna(watmdn['water_mar'].median(), inplace=True)
```

```python
#to measure RMSE, we want to only have rows that had missing values to start with

lutmeanmcar = lutmean[lutmean['mcarna'] == 1]
lutmeanmar = lutmean[lutmean['marna'] == 1]

lutmdnmcar = lutmdn[lutmdn['mcarna'] == 1]
lutmdnmar = lutmdn[lutmdn['marna'] == 1]

sugmeanmcar = sugmean[sugmean['mcarna'] == 1]
sugmeanmar = sugmean[sugmean['marna'] == 1]

sugmdnmcar = sugmdn[sugmdn['mcarna'] == 1]
sugmdnmar = sugmdn[sugmdn['marna'] == 1]

watmeanmcar = watmean[watmean['mcarna'] == 1]
watmeanmar = watmean[watmean['marna'] == 1]

watmdnmcar = watmdn[watmdn['mcarna'] == 1]
watmdnmar = watmdn[watmdn['marna'] == 1]
```

```python
#evaluate difference between imputation and truth
from sklearn.metrics import mean_squared_error
import math
import numpy as np

#lutein using mean
dat='lutein'
meth='mean'
results=[]
lutmeanmcar['sqrt']=(lutmeanmcar['lutein_zeaxanthin']-lutmeanmcar['lutein_zeaxanthin_mcar'])**2
lutmeanrmse=math.sqrt(lutmeanmcar['sqrt'].mean())
print("RMSE:",lutmeanrmse)
results.append([dat,'mcar',meth,lutmeanrmse])

lutmeanmar['sqrt']=(lutmeanmar['lutein_zeaxanthin']-lutmeanmar['lutein_zeaxanthin_mar'])**2
lutmeanrmse=math.sqrt(lutmeanmar['sqrt'].mean())
print("RMSE:",lutmeanrmse)
results.append([dat,'mar',meth,lutmeanrmse])

#lutein using median
meth='median'
lutmdnmcar['sqrt']=(lutmdnmcar['lutein_zeaxanthin']-lutmdnmcar['lutein_zeaxanthin_mcar'])**2
lutmdnrmse=math.sqrt(lutmdnmcar['sqrt'].mean())
print("RMSE:",lutmdnrmse)
results.append([dat,'mcar',meth,lutmdnrmse])

lutmdnmar['sqrt']=(lutmdnmar['lutein_zeaxanthin']-lutmdnmar['lutein_zeaxanthin_mar'])**2
lutmdnrmse=math.sqrt(lutmdnmar['sqrt'].mean())
print("RMSE:",lutmdnrmse)
results.append([dat,'mar',meth,lutmdnrmse])

#sugars using mean
dat='sugars'
meth='mean'

sugmeanmcar['sqrt']=(sugmeanmcar['sugars']-sugmeanmcar['sugars_mcar'])**2
sugmeanrmse=math.sqrt(sugmeanmcar['sqrt'].mean())
print("RMSE:",sugmeanrmse)
results.append([dat,'mcar',meth,sugmeanrmse])

sugmeanmar['sqrt']=(sugmeanmar['sugars']-sugmeanmar['sugars_mar'])**2
sugmeanrmse=math.sqrt(sugmeanmar['sqrt'].mean())
print("RMSE:",sugmeanrmse)
results.append([dat,'mar',meth,sugmeanrmse])

#sugars using median

meth='median'

sugmdnmcar['sqrt']=(sugmdnmcar['sugars']-sugmdnmcar['sugars_mcar'])**2
sugmdnrmse=math.sqrt(sugmdnmcar['sqrt'].mean())
print("RMSE:",sugmdnrmse)
results.append([dat,'mcar',meth,sugmdnrmse])

sugmdnmar['sqrt']=(sugmdnmar['sugars']-sugmdnmar['sugars_mar'])**2
sugmdnrmse=math.sqrt(sugmdnmar['sqrt'].mean())
print("RMSE:",sugmdnrmse)
results.append([dat,'mar',meth,sugmdnrmse])

#water using mean
dat='water'
meth='mean'

watmeanmcar['sqrt']=(watmeanmcar['water']-watmeanmcar['water_mcar'])**2
watmeanrmse=math.sqrt(watmeanmcar['sqrt'].mean())
print("RMSE:",watmeanrmse)
results.append([dat,'mcar',meth,watmeanrmse])

watmeanmar['sqrt']=(watmeanmar['water']-watmeanmar['water_mar'])**2
watmeanrmse=math.sqrt(watmeanmar['sqrt'].mean())
print("RMSE:",watmeanrmse)
results.append([dat,'mar',meth,watmeanrmse])

#water using median

meth='median'

watmdnmcar['sqrt']=(watmdnmcar['water']-watmdnmcar['water_mcar'])**2
watmdnrmse=math.sqrt(watmdnmcar['sqrt'].mean())
print("RMSE:",watmdnrmse)
results.append([dat,'mcar',meth,watmdnrmse])

watmdnmar['sqrt']=(watmdnmar['water']-watmdnmar['water_mar'])**2
watmdnrmse=math.sqrt(watmdnmar['sqrt'].mean())
print("RMSE:",watmdnrmse)
results.append([dat,'mar',meth,watmdnrmse])


print(results)
```
